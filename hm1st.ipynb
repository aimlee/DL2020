```
def find_x_derivative(x, y):
    x = torch.autograd.Variable(torch.Tensor([x]),requires_grad=True)
    y = torch.autograd.Variable(torch.Tensor([y]),requires_grad=True)
    func = torch.sin(torch.tanh(x)*(x*x/y) + torch.log( torch.exp(-x*x+3) + x*x*x*y) ) * torch.tanh( x*x*(torch.exp(x**9)))
    func.backward()
    return x.grad

print(find_x_derivative(1, 1))
```

```
def get_cos_sim(A, B):
    A=torch.tensor(A,dtype = float)
    B=torch.tensor(B, dtype = float)
    a_normal = torch.norm(A, dim=1)
    b_normal = torch.norm(B, dim=1,keepdim=True)
    A = A/a_normal
    B = B/b_normal
    B = B.t()
    return A.mm(B)

A = [[3, -3, 3, -3], [10, 15, -15, 22], [-3, -7, 15, 10], [12, -17, -2, 0]]
B = [[-5, -13, 11, 10, 122], [22, 88, -13, -17, -20], [20, 15, 11, 23, 11], [11, 123, 17, 7, -5]]
print(get_cos_sim(A,B))
```

```
class LinearRegression:
    def get_loss(self, preds, y):
        mse = (preds - y).pow(2).sum()
        return mse
    
    def init_weights(self, input_size, output_size):
        torch.manual_seed(0) #необходимо для воспроизводимости результатов
        self.W = torch.rand(input_size, output_size,  requires_grad=True).type(torch.DoubleTensor)
        self.b = torch.zeros(1, output_size, requires_grad=True)

    def fit(self, X, y, num_epochs=1000, lr=0.001):
        self.init_weights(X.shape[1], y.shape[1])
        metrics = []
        for _ in range(num_epochs):
            preds = self.predict(X)
            with torch.no_grad():
                dw = (preds - y).T.mm(X)
                self.W -= dw.T*lr
            metrics.append(self.get_loss(preds, y).data)
        return metrics

    def predict(self, X):
        return X@self.W
```

```
X,Y = datasets.make_regression(n_targets=3, n_features=2, noise=10, random_state=42)
X = torch.from_numpy(X).type(torch.DoubleTensor)
Y = torch.from_numpy(Y).type(torch.DoubleTensor)
model = LinearRegression()
mse = model.fit(X,Y)

print(mse)
```
